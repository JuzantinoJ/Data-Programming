<h3>Read & Save data in the CSV file into a data variable </h3>
1. Append each row into an array

```python
import csv 

with open('data.csv') as d:
    reader = csv.reader(d)
    header = next(reader) #get rid of the header line
    carpark_list = [row for row in reader]
    print(carpark_list)

```

<h3>Read & Save data in the TSV file into a data variable</h3>
1. Use - delimiter='\t'
2. reader = csv.reader(f,delimeter='\t')

```python
import csv 

with open('data.csv') as t:
    reader = csv.reader(t, delimeter('\t')) # or any same pattern
    carpark_list = [row for row in reader]
```

<h3> String Formatting </h3>

```python
#use format method on the string to insert the values into the placeholder

destination = 'Bali'
year = 2025
print('Destination to retire is {} in {}.'.format(destination,year))

#or give placeholders as variable
s = 'Destination to retire is {destination} in {year}.'
print(s.format(destination = "Bali", year = 2025))

#or use f to format string
print(f'Destination to retire is {destination} in {year}.')
```

<h3> Number Formatting </h3>

```python
# add thousands-place comma seperators when formatting numbers

print('This is the minimum sum to retire ${}'.format(250000))
print('This is the ideal sum to retire ${:,}'.format(1000000))
      
```

```python
#perform calculation and format certain numbers to decimal values

yearsToRetire = 15
amountToRetire = 250000
print('Amount to save per year to retire ${:.2f}.'.format(amountToRetire/yearsToRetire)) 

# actual value 
print('Actual value = {}'.format(7/67890.0))

# The 0 in {0:.4} refers to the 0 index in this format
# the .4 means 4 decimal values
print('7 divided by 67890 is {0:.4} in 4 decimal value'.format(7/67890.0))

# if % is provided, the result is converted into percentage
print('7 divided by 67890 is {0:.4%} in percentage'.format(7/67890.0))
```

<h3> Regular Expressions Special Characters </h3>

The Regex Cheat Sheet at http://www.rexegg.com/regex-quickstart.html provides a good reference of most regular expressions.

Char  Legend	                     Example	    Sample Match
- \d	  one digit from 0 to 9	         \d\d	         25
- \w	  ASCII letter,digit,underscore  \w\w\w	         B_1
- \s	  space, tab, newline, carriage  a\sb\sc	     a b c	
      return, vertical tab
- \S	  one character that is          \S\S\S\S	     a3d_
      not whitespace 	
- .	  any character	                 x.y	         xry or x1y	
- \	  Escapes a special character	 \.\{\}	         .{}
- +	  One or more	            \w--\w+	     A--b1_1
- {3}	  Exactly three times	         \w{3}	         Aa_
- *	  Zero or more times	         A*B*C*	         AAACCC
- ?	  Once or none	                 plurals?	     plural

```python
#use of the last regular expression “\S+@\S+” to search for email addresses

import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    #the method findall() returns a list
    x = re.findall('\S+@\S+',line) 
    if len(x) > 0:
        print(x)
```

```python
# using [a-zA-Z0-9]\S*@\S*[a-zA-Z0-9] to omit out incorrect chars
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    #the method findall() returns a list
    # [a-zA-Z0-9]\S*@\S*[a-zA-Z0-9] omits out incorrect chars
    x = re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z0-9]',line) 
    if len(x) > 0:
        print(x)
```

```python
#extract just the domain name of the email addresses
import re
hand = open('mbox-short.txt') 
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S*@(\S*[a-zA-Z0-9])', line) 
    if len(x) > 0:
        print(x[:15])
        
        #[a-zA-Z0-9]\S*@\S*[a-zA-Z0-9]'
```
